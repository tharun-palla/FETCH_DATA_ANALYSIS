### What are the top 5 brands by receipts scanned among users 21 and over?

```python
query1 = "SELECT p.BRAND AS Brand_Name, COUNT(DISTINCT t.RECEIPT_ID) AS Receipts_Scanned \
            FROM transactions t JOIN products p \
            ON t.BARCODE = p.BARCODE\
            JOIN users u \
            ON t.USER_ID = u.ID \
            WHERE CAST((JULIANDAY(CURRENT_DATE) - JULIANDAY(DATE(u.BIRTH_DATE))) / 365 AS INTEGER) > 20 \
            AND p.BRAND IS NOT NULL \
            GROUP BY p.BRAND \
            ORDER BY Receipts_Scanned DESC \
            LIMIT 5"
```

### What is the percentage of sales in the Health & Wellness category by generation?

```python
query3 = """WITH users_generation AS (
    SELECT *, 
    CASE 
            WHEN substr(u.BIRTH_DATE, 1, 4) >= '1997' THEN 'Gen Z' 
            WHEN substr(u.BIRTH_DATE, 1, 4) BETWEEN '1981' AND '1996' THEN 'Millennials' 
            WHEN substr(u.BIRTH_DATE, 1, 4) BETWEEN '1965' AND '1980' THEN 'Gen X' 
            WHEN substr(u.BIRTH_DATE, 1, 4) BETWEEN '1946' AND '1964' THEN 'Baby Boomers' 
            ELSE 'Other' 
        END AS generation FROM users u), 
    
    generation_sale as (
    SELECT u.generation, SUM(t.FINAL_SALE) final_sale
    FROM users_generation u
    INNER JOIN transactions t
    ON u.ID = t.USER_ID
    INNER JOIN products p
    ON t.BARCODE = p.BARCODE
    WHERE (p.category_1 = 'Health & Wellness' OR p.category_2 = 'Health & Wellness' 
    OR p.category_3 = 'Health & Wellness' OR p.category_4 = 'Health & Wellness')
    GROUP BY u.generation)
    
    SELECT Generation, ROUND(final_sale/(SELECT sum(final_sale) FROM generation_sale) *100, 2) AS Percent_Sales
    FROM generation_sale
    ORDER BY Percent_Sales DESC
    """
```

```python
query3_df = pd.read_sql(query3, conn)
query3_df
```

* Baby Boomers(1946-1964) contribute highest percent(46.94%) of total sales in Health & Wellness category, while Gen X(1965-1980) contribute the least of 21.88% of sales. It's surprising to see that there is no contribution from Gen Z users (1997+).

### Who are Fetch’s power users?
* Fetch's power users can be defined as users with account age more than 6 months and has atleast 15 unique receipts scanned on each month for last 3 months. This metric can give us users who are have been with Fetch for more than 6 months and are frequently using the app for the previous 3 months from current date.
* Because the last transaction recorded in the data is on 2024-09-08, It was considered as current_date in the below query

```python
query4 = """
            WITH Monthly_Receipts AS (
            SELECT USER_ID, strftime('%Y', SCAN_DATE) AS year, strftime('%M', SCAN_DATE) AS month, 
            COUNT(DISTINCT RECEIPT_ID) AS receipt_count
            FROM transactions
            WHERE SCAN_DATE >= DATE((SELECT MIN(SCAN_DATE) FROM transactions), '-3 months')
            GROUP BY user_id, year, month),
            Qualified_Users AS (
            SELECT USER_ID, receipt_count
            FROM Monthly_receipts
            WHERE receipt_count >= 15
            GROUP BY USER_ID
            HAVING COUNT(DISTINCT year || month) >=3)
            
            SELECT u.ID as user_id, q.receipt_count,
            CAST((JULIANDAY(CURRENT_DATE) - JULIANDAY(DATE(u.CREATED_DATE))) AS INTEGER) AS account_age
            FROM users u JOIN Qualified_Users q
            ON u.ID = q.USER_ID
            WHERE CAST((JULIANDAY(CURRENT_DATE) - JULIANDAY(DATE(u.CREATED_DATE))) AS INTEGER) >= 180
            ORDER BY account_age DESC"""
```

### At what percent has Fetch grown year over year?
* Fetch's growth can be defined as total number of users increased per year.

```python
query5 = """
WITH yearly_user_count AS (
    SELECT CAST(strftime('%Y', u.CREATED_DATE) AS INTEGER) AS year, COUNT(DISTINCT u.ID) AS user_count
    FROM users u
    GROUP BY year
)

SELECT this_year.year, this_year.user_count AS users_this_year, last_year.user_count AS users_last_year, ROUND(
        ((this_year.user_count - COALESCE(last_year.user_count, 0)) * 1.0 / 
        CASE WHEN last_year.user_count = 0 THEN 1 ELSE last_year.user_count END) * 100, 2) AS yoy_percentage_growth
FROM yearly_user_count this_year
LEFT JOIN yearly_user_count last_year
ON this_year.year= last_year.year + 1
ORDER BY this_year.year;

"""
```

```python
query5_df = pd.read_sql(query5, conn)
query5_df
```

* It can be observed that from 2014 - 2020 Fetch has rapidly increased it's user base. But from 2021, which is the year of covid, Fetch's growth rate slowed down considerbaly. In 2023 Fetch has observed declining user growth for the first time with 43% downturn decrease in new user acquistion. This was continued in 2024 as well but on the positive note the decline of new user quistion is better than year 2023. 2023 and 2024 indicate a significant downturn, with a drop in the user base both in absolute numbers and percentage growth, suggesting a challenging trend in user acquisition.

### Which is the leading brand in the Dips & Salsa category?
* Leading brand can be defined as a brand with most number of users that scanned their products.

```python
query6 = '''SELECT COUNT(t.USER_ID) as User_Count, p.BRAND as Brand_Name
    from transactions t LEFT JOIN users u on t.USER_ID = u.ID
    LEFT JOIN products p on t.BARCODE = p.BARCODE
    WHERE p.category_1 = 'Dips & Salsa' OR p.category_2 = 'Dips & Salsa' OR p.category_3 = 'Dips & Salsa'  OR p.category_4 = 'Dips & Salsa' 
    AND p.BRAND IS NOT NULL
    GROUP BY p.BRAND 
    ORDER BY  User_Count desc 
    Limit 1'''
```
